#### Import nödvändiga bibliotek, läser in fil

```python
import pandas as pd
import matplotlib.pyplot as plt

filename = 'dnasequenceraw.csv'

def count_dna_letters(sequence):
    #Skapar en dictionary som räknar DNA-bokstäver
    dna_count = {'A': 0, 'T': 0, 'C': 0, 'G': 0}
    
    #Räknar förekomsten av varje bokstav i den specifika sekvensen
    for letter in sequence.upper():#upper ändrar till stora bokstäver
        if letter in dna_count:
            dna_count[letter] += 1
    
    return dna_count
```

#### Plotta frekvenser/ Stapeldiagram över frekvenser med biblioteket matplotlib

```python
def plot_dna_letter_frequency(sequence_id, dna_count):
    #Plotta frekvenserna för varje enskild DNA-bokstav
    letters = list(dna_count.keys())
    counts = list(dna_count.values())
    
    plt.figure(figsize=(8, 6))
    plt.bar(letters, counts, color=['red', 'blue', 'green', 'orange'])
    plt.title(f"DNA Letter Frequency for {sequence_id}")
    plt.xlabel("DNA Letters")
    plt.ylabel("Count")
    plt.show()
```

```python
def process_dna_file(filename):
    df = pd.read_csv(filename, encoding='utf-8')#Läser in dna filen
    df.columns = df.columns.str.strip().str.lower()#Justerar kolumnnamnen

    #Bearbetar varje sekvens
    for index, row in df.iterrows():
        sequence_id = row['id']
        sequence = row['seq']
        
        #Räknar DNA-bokstäver
        dna_count = count_dna_letters(sequence)
        
        #Skriver ut det slutliga resultatet 
        print(f"Sequence ID: {sequence_id}")
        print(f"DNA letter counts: {dna_count}")
        
        #Plotta frekvenserna för DNA-bokstäverna
        plot_dna_letter_frequency(sequence_id, dna_count)

#Anropar funktionen som bearbetar filen
process_dna_file(filename)
```
